#### Relu - Rectified linear unit 
` Relu - This function returns 0 if it takes a negative argument, in the case of a positive argument, the function returns the number itself. `

<img src="Photo/relu2.png">

```python
def relu(x):
    return max(0, x)
```

```python
x = 10
relu(x)
```

```python
x = -0.5
relu(x)
```

#### Sigmoid
`Sigmoid - The sigmoid function is a mathematical function that has a characteristic that can take any real value and map it to between 0 to 1  `

<img src="Photo/sigmoid.png">

```python
from math import exp
```

```python
def sigmoid(x):
    return 1/(1 + math.exp(-x))
```

```python
x = 1
sigmoid(x)
```

```python
x = -1
sigmoid(x)
```

```python
x = 10
sigmoid(x)
```

```python
x = -10
sigmoid(x)
```

#### Softmax
`Softmax - is a function that turns logits (sets of numbers) into probabilities, and the sum of the latter is equal to one. The function outputs as a result a vector representing the probability distributions of the list of potential outcomes.  `

` Softmax works by probability. At the end, when the neuron exits the network, the sum of probabilities is equal to 1. `

` The function turns the input [2.0, 1.0, 0.1] into probabilities [0.7, 0.2, 0.1] (0.7+0.2+0.1=1) `

<img src="Photo/softmax.png">

```python
def softmax1(x):
    #calculate each probability
    p1 = exp(x[0]) / (exp(x[0]) + exp(x[1]) + exp(x[2]) + exp(x[3]))
    p2 = exp(x[1]) / (exp(x[0]) + exp(x[1]) + exp(x[2]) + exp(x[3]))
    p3 = exp(x[2]) / (exp(x[0]) + exp(x[1]) + exp(x[2]) + exp(x[3]))
    p4 = exp(x[3]) / (exp(x[0]) + exp(x[1]) + exp(x[2]) + exp(x[3]))
    
    # report sum of probabilities
    print(p1+p2+p3+p4)
    
    # report probabilities
    return [p1, p2, p3, p4]
```

```python
x = [3.2, 1.3, 0.2, 0.8] #Logits
softmax1(x)
```
