```python
import numpy as np

x1=np.array([2,4,3,1,2])
x2=np.array([1,3,4,2,4])
x3=np.array([3,1,2,3,1])

y=np.array([24,26,31,25,25])
```

```python
# random weight and bias
w1=-2
w2=3
w3=2
b1=1.5

w4=-2.4
w5=3.2
w6=1.5
b2=-1.8

w7=2.6
w8=3.2
b3=2
```

```python
for i in range(500):
    for x1_i, x2_i, x3_i, y_i in zip(x1,x2,x3,y):
        
        y1 = w1 * x1_i + w2 * x2_i + w3 * x3_i + b1
        y2 = w4 * x1_i + w5 * x2_i + w6 * x3_i + b2
        y3 = w7 * y1 + w8 * y2 + b3
        
        w7 = w7 - 0.001 * (2 * y1 * (y3 - y_i)) # 0.001 = learning rate parametr
        w8 = w8 - 0.001 * (2 * y2 * (y3 - y_i))
        b3 = b3 - 0.001 * (2 * (y3 - y_i))
        
        w1 = w1 - 0.001 * (2* w7 * x1_i * (y3 - y_i))
        w2 = w2 - 0.001 * (2* w7 * x2_i * (y3 - y_i))
        w3 = w3 - 0.001 * (2* w7 * x3_i * (y3 - y_i))
        b1 = b1 - 0.001 * (2 * w7 * (y3 - y_i))
        
        w4 = w4 - 0.001 * (2* w8 * x1_i * (y3 - y_i))
        w5 = w5 - 0.001 * (2* w8 * x2_i * (y3 - y_i))
        w6 = w6 - 0.001 * (2* w8 * x3_i * (y3 - y_i))
        b2 = b2 - 0.001 * (2 * w8 * (y3 - y_i))
        
    if i%10==0:
        print("Weight 1 = ", w1, " Weight 2 = ", w2, " Weight 3 = ", w2, " Bias 1 = ",b1)
        print("Weight 4 = ", w4, " Weight 5 = ", w5, " Weight 6 = ", w6, " Bias 2 = ",b2)
        print("Weight 7 = ", w7, " Weight 8 = ", w8, " Bias 3 = ",b3)
        print('----------------------------------')
```
