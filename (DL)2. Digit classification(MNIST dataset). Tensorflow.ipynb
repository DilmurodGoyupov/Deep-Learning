```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

# <center> Reading MNIST dataset

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

# <center> 1 - Normalize the dataset

* we bring the dataset values between 0 and 1. To do this, we divide the dataset by its largest value, which is 255

```python
x_train = x_train / 255
x_test = x_test / 255
```

```python
x_train[0]
```

```python
x_test[0]
```

# <center> 2 - Flatten the dataset

* We feed the images to the neural network with a size of 28 x 28 by 60000. The input layer has 28*28=784 neurons, each neuron is 1 pixel

```python
x_train2 = x_train.reshape(60000, 28*28)
x_test2 = x_test.reshape(10000, 28*28)
```

```python
x_train2.shape
```

```python
x_test2.shape
```

```python
x_train2[0]
```

```python
x_train2[0].shape
```

# <center> 3 - Create model

```python
model = keras.Sequential([
    keras.layers.Dense(16, input_shape=(784, ), activation='relu'),   #28*28=784
    keras.layers.Dense(16, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
    
])
```

## <center> View of our architecture



```python
model.summary()
```

# <center> 4 - Give the parameters of the model

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics='accuracy')
```

# <center> 5 - Train the model

```python
model.fit(x_train2,y_train,epochs=10)
```

# <center> 6 - Model checking

```python
model.evaluate(x_test2,y_test)
# our model works with 95% accuracy
```

# <center> 7 - Predict 

```python
predict_all = model.predict(x_test2)
```

```python
predict_all
```

```python
# we predict the 10th picture
# prediction value for each number in the matrix
predict_all[10]
```

```python
# the biggest is our prediction
np.argmax(predict_all[10])
```

```python
plt.imshow(x_test[10],cmap='gray')
```

# <center> 8 - View in heatmap

```python
predict_all_value = [np.argmax(i) for i in predict_all]
```

```python
matrix = tf.math.confusion_matrix(y_test, predict_all_value)
```

```python
plt.figure(figsize=(6,6))
sn.heatmap(matrix, annot=True, fmt="d")
plt.xlabel("Real")
plt.ylabel("Predict")
plt.show()
```

# <center> 9 - Save model

```python
model.save("mnist_recognition1",save_format="h5")
```
