* ### A convolutional neural network is a special architecture of artificial neural networks proposed by Yann LeCun in 1988 to efficiently detect patterns and is part of deep learning technologies. 

Its name is so called because of the existence of a convolution operation, the essence of which is that each image piece is multiplied by the element of the convolution matrix (kernel), and the result is summarized, and the output is written to a similar position of the image.
